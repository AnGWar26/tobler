```python
def w_from_network(network, distance=45, num_pois=500):
    neighbors = network.nearest_pois(distance, "neighbors", num_pois=500, include_poi_ids=True, )
    cols = neighbors.columns.astype(str)
    c = [col for col in cols if col.startswith('poi')]
    neighbors=neighbors[c]
    neighbors = neighbors.dropna(how='all')
    neighbors = neighbors.T
    a={k:v.dropna().astype(int) for k, v in neighbors.reset_index(drop=True).items()}   
    return a
```
